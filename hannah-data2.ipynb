# The Metric:
The metric I've decided to use to determine the best neighborhood in Pittsburgh is the amount of traffic in each neighborhood.  The less traffic counts the neighborhood has the "better" it is.

```python
import pandas as pd
import numpy as np
```

```python
pgh_traffic_count = pd.read_csv("https://data.wprdc.org/datastore/dump/6dfd4f8f-cbf5-4917-a5eb-fd07f4403167",
                           index_col="id", 
                           parse_dates=True)

pgh_traffic_count.head()
```

```python
traffic = pgh_traffic_count[['neighborhood','average_daily_car_traffic']]
traffic.head(10)
```

The code ran below calculates the total number of traffic counts for each neighborhood in the dataset:

```python
traffic_total = traffic.groupby("neighborhood").sum()
traffic_order = traffic_total.sort_values('average_daily_car_traffic',ascending=True)
traffic_order.head(25)
```

The neighborhoods with 0 average daily car tarffic must not of had sufficient data because zero car traffic isn't possible in any neighborhood.  Therefore, I will not be including those in the data to determine the top neighborhoods with the least amount of traffic.

```python
import pandas as pd
import geopandas
%matplotlib inline
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") 
neighborhoods.plot() 
```

```python
traffic_map = neighborhoods.merge(traffic_order, how='left', left_on='hood', right_on='neighborhood')
traffic_map[['hood','average_daily_car_traffic','geometry']].head(10)
```

```python
traffic_map.plot(column='average_daily_car_traffic')
```

```python
traffic_map.plot(column='average_daily_car_traffic', 
               cmap='BuPu',             
               edgecolor="white",        
               legend=True,              
               legend_kwds={'label': "Traffic Count"},
               figsize=(15, 10),        
               missing_kwds={"color": "lightgrey"}
               )
```
